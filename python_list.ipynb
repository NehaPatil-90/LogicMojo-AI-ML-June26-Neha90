```python
my_list = (1,2,3,4,5,6,7) 

sliced_list = my_list[1:6:2]
print(sliced_list)

sliced_list = my_list[:6:2]
print(sliced_list)

sliced_list = my_list[::2]
print(sliced_list)

sliced_list = my_list[-7:-1]
print(sliced_list)

sliced_list = my_list[::-1]
print(sliced_list)
```
