```python
numbers = [1,2,3,4]

numbers.append(5)
print(numbers)
numbers.extend([6,7])
print(numbers)
numbers.insert(1,0)
print(numbers)
numbers.remove(2)
print(numbers)


list1 = [1,2,3,4]
list2 = ["Alice","Bob","Kat"]

for index, num in enumerate(list1):
    print(f"Index {index}: {num}")

for list1,list2 in zip(list1,list2):
    print(f"{list1}:{list2}")

list1 = [1,2] *3
print(f"Repeated: {list1}")

items= ['a','b','c','d','e']
print(f"Original: {items}")
items[1] ='B'
print(f"Modified: {items}")
items[1:3]=['B','C']
print(f"Modified: {items}")
```
